# Retrieve data

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Spotify_Youtube.csv')
```

```python
droppedcolumns = ["Url_spotify","Uri","Url_youtube","Description"] # Data irrelevant to the content of the song
for x in droppedcolumns:
    df = df.drop(labels = x , axis=1)
```

```python
# Removing rows with missing values
df = df.dropna()

# Removing songs that exceed 10 minutes, getting rid of compilations and ASMR
df = df.drop(df[df.Duration_ms > 600000].index)
```

```python
# Semi-Manual Check for audiobooks and other non-songs

songs_removed1 = []
number1 = []

blacklist = ["Folge","Teil","Meditation","Hz","These Words"] 


for x in range(0,len(df.index)):
    for y in blacklist:
        if y in df.iloc[x]['Album'] or y in df.iloc[x]['Track'] or y in df.iloc[x]['Title']:
            songs_removed1.append(str(df.iloc[x][2]))
            number1.append(x)

bchannels = ["Rain Fruits Sounds", "Ocean Sounds","Ocean Waves For Sleep","White Noise for Babies", "White Noise Baby Sleep", "Sleepy Penguin"]

for x in range(0,len(df.index)):
    for y in bchannels:
        if y in df.iloc[x]["Artist"]:
            songs_removed1.append(str(df.iloc[x][2]))
            number1.append(x)


number1 = list(dict.fromkeys(number1))
number1.sort()

# Drop here
number = df.index[number1]
df = df.drop(number)

# Final csv
df.to_csv("cleaneddf.csv")

#print total null rows
print(df.isnull().sum())
```
